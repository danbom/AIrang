```
txt = "const path = require('path');\nconst express = require('express');\nconst app = express();\nvar portNum = 3000;\n\napp.use("
print(txt + "gpt")
```

```
start = time.time()
response = openai.Completion.create(
    engine="davinci",
    prompt=txt,
    temperature=0.7,
    # frequency_penalty=0.5,
    max_tokens=180,         # 3 tk = 1 al
    top_p=1
)

data = response['choices'][0]['text']
print(data)
print("time :", time.time() - start)
print("length: " + str(len(data)) + "\nactual len: " + str(len(re.findall("[\S]", data))) + "\n")
```
